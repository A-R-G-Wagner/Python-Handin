## Ex 1 - Download the data

```python
import csv
import requests as req
import requests
import io
import pandas as pd
import sklearn
import sklearn.linear_model
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
#Solution TO FILE
Data = 'https://think.cs.vt.edu/corgis/datasets/csv/cars/cars.csv'
File = 'files/carsCSV.csv'

def download(url,file):
    try:
        r = req.get(url)
        r.raise_for_status()
        with open(file, 'w') as f:
            f.write(r.text)
    except req.exceptions.HTTPError as e:
        print(e.response.text)
    except FileNotFoundError:
        print("File not found, jerk!")
        
download(Data,File)
```

```python
#Solution FROM NET
url = "https://think.cs.vt.edu/corgis/datasets/csv/cars/cars.csv"
s = requests.get(url).content
df = pd.read_csv(io.StringIO(s.decode('utf-8')))
df.head(0)
```

## Feature engineering

```python
#Sorterer på kolonne værdier
df_honda_gasoline = df.loc[(df['Identification.Make']=='Honda') & (df['Fuel Information.Fuel Type']=='Gasoline')]
df_honda_gasoline
```

```python
#Fjerner outliers
df_honda_gasoline_clean = df_honda_gasoline.loc[(df_honda_gasoline['Engine Information.Engine Statistics.Horsepower'] < df_honda_gasoline['Engine Information.Engine Statistics.Horsepower'].quantile(0.90)) & 
(df_honda_gasoline['Fuel Information.Highway mpg'] < df_honda_gasoline['Fuel Information.Highway mpg'].quantile(0.90))]
df_honda_gasoline_clean
```

```python
#reducerer til kun at vise 2 kolonner
df_honda_gasoline_clean_reduced = df_honda_gasoline_clean[['Fuel Information.Highway mpg', 'Engine Information.Engine Statistics.Horsepower']]
df_honda_gasoline_clean_reduced
```

```python
#normalisering af værdier

#Using sklearning.preprocessing
#scaler = preprocessing.MinMaxScaler()
#names = df_honda_gasoline_clean_reduced.columns
#d = scaler.fit_transform(df_honda_gasoline_clean_reduced)
#scaled_df = pd.DataFrame(d, columns=names)

##Mathematical expression
df_honda_gasoline_clean_reduced_normalized = ( (df_honda_gasoline_clean_reduced-df_honda_gasoline_clean_reduced.min()) / (df_honda_gasoline_clean_reduced.max() - df_honda_gasoline_clean_reduced.min()) )

#scaled_df.head()
df_honda_gasoline_clean_reduced_normalized
```

## Linear regression

```python
#Plottet med hestekrafter i x-aksen og miles per galon i y-aksen
df_honda_gasoline_clean_reduced_normalized.plot.scatter(x=1,y=0)

#Best Guess
plt.plot([1,0], [0.1,0.9])
```

```python
xs = df_honda_gasoline_clean_reduced_normalized['Fuel Information.Highway mpg']
ys = df_honda_gasoline_clean_reduced_normalized['Engine Information.Engine Statistics.Horsepower']

xs_reshape = np.array(xs).reshape(-1, 1)

model = sklearn.linear_model.LinearRegression()
refitted_model = model.fit(xs_reshape, ys)


#Koeficienten indikerer stigningen eller faldet på modellen, per enhed.
#(+)indikere stigning (-)indikere fald, jo tættere på nul, jo mere vandret.
refitted_model.coef_
```

```python
#Intercept er hvor modellen skærer y-aksen, altså hvor x-koordinatet er 0.
refitted_model.intercept_
```

```python
#Her beregnes y-koordinatet, når x-koordinatet er 1.
predicted = model.predict(xs_reshape)
target = 1
hpPrediction = model.predict([[target]])
print('{}'.format(hpPrediction[0]))
```

```python
df_honda_gasoline_clean_reduced_normalized.plot.scatter(x=1, y=0)
plt.plot([0,target], [1.0229952323711924,hpPrediction[0]])
```
